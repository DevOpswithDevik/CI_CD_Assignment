1. Simple Application Creation
Goal: Establish the source code and define the build command.

```

name: Node.js CI Build

# 1. When should this workflow run?
on:
  push:
    branches: [ main ] # Trigger the workflow on every push to the 'main' branch
  pull_request:
    branches: [ main ] # Also trigger on pull requests

# 2. What jobs should run?
jobs:
  build:
    runs-on: ubuntu-latest # Specify the runner environment

    steps:
      # Step 1: Check out the code
      - uses: actions/checkout@v4
        with:
          fetch-depth: 0 # Fetch all history for better analysis

      # Step 2: Set up Node.js environment
      - name: Set up Node.js
        uses: actions/setup-node@v4
        with:
          node-version: '20.x' # Use Node.js version 20

      # Step 3: Install dependencies
      - name: Install Dependencies
        run: npm install

      # NEW STEP: Run Automated Tests
      - name: Run Automated Tests
        run: npm test

      # Step 4: Run the build command defined in package.json
      - name: Run Build
        run: npm run build

      # NEW STEP: Archive and Upload the Built Code as a Deployable Artifact
      - name: Archive Production Artifact
        uses: actions/upload-artifact@v4 # Action to upload files
        with:
           name: web-app-package-${{ github.sha }} # Unique name for the package
           path: .
              # Exclude the heavy node_modules directory from the final deployable artifact
              # Note: For production use, you would typically run a dedicated build step
              # that includes only the necessary files.
           retention-days: 5  


```
